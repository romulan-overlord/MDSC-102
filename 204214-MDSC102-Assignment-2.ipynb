arange() takes a start point(inclusive) and an end point(exclusive) and a step size.
It returns an array of elements starting at the starting point, each element incrementing by step size till it reaches a step before the end point.

```python
x = np.zeros((2,3))

np.arange(5, 55,2)
# np.ones_like(x)
```

linspace() takes start and end points and an element count, n. Generates an array of n elements, first elemet is start point, last element is end point, each element is evenly spaced.

```python
np.linspace(1,2,10, retstep=True)
```

fromfunction() takes a function and a tuple as parameters.
takes the  tuple and starting from (0,0), calls the function passed using the tuple iteration till it reaches the actual tuple.  

```python
mat1 = np.fromfunction(lambda i,j: 2*i+j, (4,3))

n = 3
np.fromfunction(lambda x: (x % n == 0) * 1, (20,))

mat1.dtype
```

Now we try to generate a bunch of points, pass those co-ordinates to a function and plot it.

```python
f = lambda x,y : 0.3*x**3 + y**2 + 2*x*y -6*x - 3*y +4
x = np.arange(-15,15,0.1)   #starts at -15. goes till 15, step size = 0.1
y = np.arange(-15,15,0.1)

fig = plt.figure(figsize = (16,8))
X, Y = np.meshgrid(x,y)
Z = f(X,Y)
N = 50
cp = plt.contour(X,Y,Z,N,cmap='jet')
plt.clabel(cp, inline=1, fontsize=8)
plt.axis('scaled')
plt.show()
# print(x)
```

```python
x = np.array([1,4,9,16])
print(np.mean(x))
print(np.median(x))

print(np.average(x, weights = [1,2,3,4]))

print(np.std(x))

print(np.sqrt(np.var(x)))

np.cov(x)
```
