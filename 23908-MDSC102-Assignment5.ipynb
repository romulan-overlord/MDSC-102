```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
names = 'Class,Alcohol,Malic Acid,Ash,Alcalinity of ash,Magnesium,Total Phenols,Flavanoids,Nonflavanoid phenols,Proanthocyanins,Color intensity,Hue,OD280/OD315 of diluted wines,Proline'
names = names.split(',')
df = pd.read_csv('./../wine/wine.data', names=names)
df
```

```python
for index, column in enumerate(df):
    print(index, column)
```

```python
plt.figure(figsize=(20,30))
for index, column in enumerate(df):
    if index > 0 and index < 13:
        plt.subplot(4,3,index)
        sns.boxplot(df, x='Class', y=column, hue='Class', palette='bright')
        plt.grid()

plt.show()
```

```python
sns.histplot(df[df['Class'] == 3]['Alcalinity of ash'], kde=True)
```

## Inferences

Alcohol mean varies across the three classes. Class 1 and 3 look normal, while class 2 feels left skewed if we don't account for the outliers.

Malic acid has similar means in class 1 and 2, but its value has high variance in class 3.

All classes have similar Ash levels, class 3 is right skewed.

Alcalinity of ash: class 1 ia right skewed, class 2 is normal, class 3 is also right skewed.

Magnesium: class 1 and 2 are right skewed, class 3 slightly less so. All have similar means.

Total phenols varies in the three classes. class 1 is normal, with a high mean, class 2 is more or less normal but with more spread, class 3 is left skewed with a low mean.

Flavanoids: all classes have different means. Class 1 is highest, class 2 is in the middle, class 3 is the lowest, while being right skewed.

Nonflavanoid Phenols: class 1 looks normal with a few outliers. class 2 has high variance. class 3 is left skewed.

Proanthocyanins: class 1 has a slightly higher mean than class 2. class 3 has the lowest mean. 

Color Intensity: class i is right skewed, class 2 has a low mean and variance, class 3 is also right skewed.

Hue: class 1 and class 2 have similar means, class 1 look slightly left skewed, class 3 has a lower mean, and is right skewed.

OD280/OD315 of diluted wines: class 2 is left skewed, class 3 is right skewed. class 1 looks normal.






```python
plt.figure(figsize=(20,50))
i = 1
for column in df:
    plt.subplot(14,3,i)
    sns.histplot(df[df['Class'] == 1][column], kde=True)
    plt.grid()
    i += 1

    plt.subplot(14,3,i)
    sns.histplot(df[df['Class'] == 2][column], kde=True)
    plt.grid()
    i += 1

    plt.subplot(14,3,i)
    sns.histplot(df[df['Class'] == 3][column], kde=True)
    plt.grid()
    i += 1
plt.show()
```

## Verification

Alcohol mean varies across the three classes. Class 1 and 3 look normal, so does class 2.

Malic acid has similar means in class 1 and 2, but its value has high variance in class 3. Class 1 and 2 are right skewed.

All classes have similar Ash levels, class 3 is right skewed. so is class 1.

Alcalinity of ash: class 1 is right skewed, class 2 is normal, class 3 is also right skewed.

Magnesium: class 1 and 2 are right skewed, class 3 slightly less so. All have similar means.

Total phenols varies in the three classes. class 1 is normal, with a high mean, class 2 is more or less normal but with more spread, class 3 is right skewed with a low mean.

Flavanoids: all classes have different means. Class 1 is highest, class 2 is in the middle, class 3 is the lowest, while being right skewed. class 2 is also right skewed.

Nonflavanoid Phenols: class 1 looks right skewed. class 2 has high variance. class 3 is left skewed.

Proanthocyanins: class 1 has a slightly higher mean than class 2. class 3 has the lowest mean. All are right skewed.

Color Intensity: class 1 is right skewed, class 2 has a low mean and variance, class 2 is also right skewed.

Hue: class 1 and class 2 have similar means, class 1 looks slightly left skewed, class 3 has a lower mean, and is right skewed, along with class 2.

OD280/OD315 of diluted wines: class 2 is left skewed, class 3 is right skewed. class 1 looks normal.
