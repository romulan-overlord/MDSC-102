```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
names = 'Class,Alcohol,Malic Acid,Ash,Alcalinity of ash,Magnesium,Total Phenols,Flavanoids,Nonflavanoid phenols,Proanthocyanins,Color intensity,Hue,OD280/OD315 of diluted wines,Proline'
names = names.split(',')
df = pd.read_csv('./../wine/wine.data', names=names)
```

```python
# sns.scatterplot(df, x='Alcohol', y='Ash', hue='Class')
plt.figure(figsize=(20,20))
for index, column in enumerate(df):
    if index > 1:
        plt.subplot(4,3,index-1)
        sns.scatterplot(df, x='Alcohol', y=column, hue='Class', palette='bright')

plt.show()
```

## Inferences

Alcohol level can be used to classify class 2 wine from class 1 wine with decent reliability. Class 3 wine cannot be identified using alcohol level alone.

Class 1 and 2 have lower levels of malic acid, while class 3 is more spread out.

Class 1 and 3 have similar Ash values while class 2 has more variance.

Class 1 has lower alcalinity of ash, class 3 has slightly higher alcalinity.

All classes have magnesium levels between 80 and 120, but class 2 has some outliers.

Class 3 has significantly lesser phenols than class 1 and 2. This can be used to classify class 3 wine from the others.

Class 3 also has much lesser flavanoids. plotted against alcohol content, flavanoids and phenols divide the types with a comfortable margin.

Hue and OD280/OD315 of diluted wines also divide the classes in a similar fashion, with class 3 having lower values than the other two.

Class 1 wine has higher Proline than class 2 and 3.
