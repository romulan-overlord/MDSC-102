### Plot the following distributions


$$
B(n, p) = \binom{n}{x}p^{x} * (1-p)^{n-x}
\\ n = 30, p = 0.2
$$


$$
Poi(\lambda) = e^{-\lambda} \frac{{\lambda}^{x}}{x!}
\\ \lambda = (0.01, 1, 5)
$$


$$
Exp(\lambda) = \lambda  e^{-{\lambda}x}
$$


$$

$$


```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

## Binomial

```python
binomial = lambda n,p,x : math.comb(n,x) * p**x  * (1-p)**(n-x)

xpoints = np.arange(0,31)
ypoints = np.array([binomial(30,0.2,x) for x in xpoints])

plt.xlabel('x-points')
plt.ylabel('y-points')
plt.title('Binomial Distribution')

plt.bar(xpoints, ypoints)

plt.show()
```

## Poisson

```python
poisson = lambda lam, x : np.exp(-lam) * (lam**x) / math.factorial(x)

lam = 5
xpoints = np.arange(0,15)
ypoints = np.array([poisson(lam,x) for x in xpoints])

plt.xlabel('x-points')
plt.ylabel('y-points')
plt.title('Poisson Distribution')

plt.stem(xpoints, ypoints, linefmt='black', label='Lambda = 5')

lam = 0.01
ypoints = np.array([poisson(lam,x) for x in xpoints])
plt.stem(xpoints, ypoints, linefmt='blue', label='Lambda = 0.01')

lam = 1
ypoints = np.array([poisson(lam,x) for x in xpoints])
plt.stem(xpoints, ypoints, linefmt='green', label='Lambda = 1')

plt.legend()
plt.show()
```

## Exponential

```python
exponential = lambda lam, x : lam * np.exp(-lam * x)

lam = 0.01
xpoints = np.linspace(0,5,100)
ypoints = np.array([exponential(lam,x) for x in xpoints])
plt.plot(xpoints,ypoints, c='black', label=f'lambda = {lam}')

lam = 1
ypoints = np.array([exponential(lam,x) for x in xpoints])
plt.plot(xpoints,ypoints, c='red', label=f'lambda = {lam}')

lam = 5
ypoints = np.array([exponential(lam,x) for x in xpoints])
plt.plot(xpoints,ypoints, c='green', label=f'lambda = {lam}')

plt.xlabel('x-points')
plt.ylabel('y-points')
plt.title('Exponential Distribution')
plt.legend()
plt.show()
```

### Defining functions:

```python
def plot_binomial(n,p):
    xpoints = np.arange(0,n+1)
    ypoints = np.array([binomial(n,p,x) for x in xpoints])
    plt.plot(xpoints, ypoints, c='hotpink', label='binomial')

def plot_poisson(lam, n):
    xpoints = np.arange(0,n+1)
    ypoints = np.array([poisson(lam,x) for x in xpoints])
    plt.stem(xpoints, ypoints, label='poisson')

def plot_exponential(lam, x):
    xpoints = np.linspace(0,10,100)
    ypoints = np.array([exponential(lam,x) for x in xpoints])
    plt.plot(xpoints,ypoints, label='exponential')
    

plot_binomial(20,0.2)
plot_poisson(5,20)
plot_exponential(1,20)

plt.title('All Plots')
plt.legend()

plt.show()


```
