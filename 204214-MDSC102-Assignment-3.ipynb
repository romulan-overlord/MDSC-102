### Plot the following distributions


$$
B(n, p) = \binom{n}{x}p^{x} * (1-p)^{n-x}
\\ n = 30, p = 0.2
$$


$$
Poi(\lambda) = e^{-\lambda} \frac{{\lambda}^{x}}{x!}
\\ \lambda = (0.01, 1, 5)
$$


$$
Exp(\lambda) = \lambda  e^{-{\lambda}x}
$$


$$
N(\mu, {\sigma}^2) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{1}{2}({\frac{x - \mu}{\sigma}})^2}
$$


$$
Gamma(\alpha, \lambda) = \frac{\lambda^\alpha}{\Gamma(\alpha)} x^{\alpha - 1} e^{-\lambda x}

\\
where:
\\
\Gamma(\alpha) = \int_{0}^{\infty}x^{\alpha - 1} e^{-x}\,dx
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

## Binomial

```python
binomial = lambda n,p,x : math.comb(n,x) * p**x  * (1-p)**(n-x)

xpoints = np.arange(0,31)
ypoints = np.array([binomial(30,0.2,x) for x in xpoints])

plt.xlabel('x-points')
plt.ylabel('y-points')
plt.title('Binomial Distribution')

plt.bar(xpoints, ypoints)

plt.show()
```

## Poisson

```python
poisson = lambda lam, x : np.exp(-lam) * (lam**x) / math.factorial(x)

lam = 5
xpoints = np.arange(0,15)
ypoints = np.array([poisson(lam,x) for x in xpoints])

plt.xlabel('x-points')
plt.ylabel('y-points')
plt.title('Poisson Distribution')

plt.stem(xpoints, ypoints, linefmt='black', label='Lambda = 5')

lam = 0.01
ypoints = np.array([poisson(lam,x) for x in xpoints])
plt.stem(xpoints, ypoints, linefmt='blue', label='Lambda = 0.01')

lam = 1
ypoints = np.array([poisson(lam,x) for x in xpoints])
plt.stem(xpoints, ypoints, linefmt='green', label='Lambda = 1')

plt.legend()
plt.show()
```

## Exponential

```python
exponential = lambda lam, x : lam * np.exp(-lam * x)

lam = 0.01
xpoints = np.linspace(0,5,100)
ypoints = np.array([exponential(lam,x) for x in xpoints])
plt.plot(xpoints,ypoints, c='black', label=f'lambda = {lam}')

lam = 1
ypoints = np.array([exponential(lam,x) for x in xpoints])
plt.plot(xpoints,ypoints, c='red', label=f'lambda = {lam}')

lam = 5
ypoints = np.array([exponential(lam,x) for x in xpoints])
plt.plot(xpoints,ypoints, c='green', label=f'lambda = {lam}')

plt.xlabel('x-points')
plt.ylabel('y-points')
plt.title('Exponential Distribution')
plt.legend()
plt.show()
```

## Normal

```python
normal = lambda mu, sigma, x: (np.exp(-0.5 * ((x - mu) / sigma)**2) / math.sqrt(2*np.pi*sigma**2))

mu = 0
sigma = 1
xpoints = np.linspace(-10,10,100)
ypoints = np.array([normal(mu, sigma, x) for x in xpoints])
plt.plot(xpoints,ypoints, c='orange', label=f'mu = {mu}, sigma = {sigma}')

sigma = 2
ypoints = np.array([normal(mu, sigma, x) for x in xpoints])
plt.plot(xpoints,ypoints, c='red', label=f'mu = {mu}, sigma = {sigma}')

sigma = 4
ypoints = np.array([normal(mu, sigma, x) for x in xpoints])
plt.plot(xpoints,ypoints, c='purple', label=f'mu = {mu}, sigma = {sigma}')

plt.xlabel('x-points')
plt.ylabel('y-points')
plt.title('Normal Distribution')
plt.legend()
plt.show()
```

## Gamma

```python
def gamma_function(alpha):
    if alpha <= 0: return None
    if alpha == 0.5: return np.pi
    if alpha <= 2: return 1
    return (alpha-1) * gamma_function(alpha - 1)

gamma = lambda alpha, lam, x : ((lam**alpha) * x**(alpha - 1) * np.exp(-lam * x))/gamma_function(alpha)

alpha = 1
lam = 1
xpoints = np.linspace(0,15,100)
ypoints = np.array([gamma(alpha, lam, x) for x in xpoints])
plt.plot(xpoints, ypoints, c='red', label=f'alpha = {alpha}, lambda = {lam}')

alpha = 2
ypoints = np.array([gamma(alpha, lam, x) for x in xpoints])
plt.plot(xpoints, ypoints, c='orange', label=f'alpha = {alpha}, lambda = {lam}')

alpha = 4
ypoints = np.array([gamma(alpha, lam, x) for x in xpoints])
plt.plot(xpoints, ypoints, c='yellow', label=f'alpha = {alpha}, lambda = {lam}')

plt.xlabel('x-points')
plt.ylabel('y-points')
plt.title('Gamma Distribution')
plt.legend()
plt.show()
```

### Defining functions:

```python
def plot_binomial(n,p):
    xpoints = np.arange(0,n+1)
    ypoints = np.array([binomial(n,p,x) for x in xpoints])
    plt.plot(xpoints, ypoints, c='hotpink', label='binomial')

def plot_poisson(lam, n):
    xpoints = np.arange(0,n+1)
    ypoints = np.array([poisson(lam,x) for x in xpoints])
    plt.stem(xpoints, ypoints, label='poisson')

def plot_exponential(lam, x):
    xpoints = np.linspace(0,10,100)
    ypoints = np.array([exponential(lam,x) for x in xpoints])
    plt.plot(xpoints,ypoints, label='exponential')
    

plot_binomial(20,0.2)
plot_poisson(5,20)
plot_exponential(1,20)

plt.title('All Plots')
plt.legend()

plt.show()


```
